```python
L=[1,7,9,16,1,9,56,98,11]
```

```python
def Bubble_sort(L):
    i=0
    while i<len(L):
        for j in range(len(L)-1):
            if L[j]>L[j+1]:
                copie=L[j+1]
                L[j+1]=L[j]
                L[j]=copie
            print(L)
        i+=1
    return L
```

```python
Bubble_sort(L)
```
